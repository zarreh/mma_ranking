```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pandas as pd
import seaborn as sn
import bar_chart_race as bcr
```

```python
BASE_PATH = "../UFC_ranking"
FILE_NAME = "rank_long"
DATE_TIME_TAG = "20240322_1203"
```

```python
%pwd
```

```python
file_path = f"{BASE_PATH}/{FILE_NAME}_{DATE_TIME_TAG}.csv"
print(file_path)
df = pd.read_csv(file_path)
df = df[df.wight_class_rank != "wight_class_rank"]
numeric_columns = [
    "page_rank",
    "perf_rat",
    "n_fights",
    "adj_p_rank",
    "wight_class_rank",
]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)
df["run_date"] = pd.to_datetime(df["run_date"])  # .dt.date
```

```python
fighter_list = (
    df[(df.weight == "Lightweight") & (df.wight_class_rank <= 15)]
    .fighter.unique()
    .tolist()
)
df_plot = df[
    (df.fighter.isin(fighter_list)) & (df.weight == "Lightweight")
].reset_index()
```

```python
# df_plot = df[(df.weight == "Lightweight") & (df.wight_class_rank <= 10)].sort_values(
#     by=["run_date", "wight_class_rank"]
# )
df_plot = df_plot[["run_date", "fighter", "adj_p_rank"]].drop_duplicates()
```

```python
df_pivot = df_plot.pivot(index="run_date", columns="fighter", values="adj_p_rank")
df_pivot
```

```python
bcr.bar_chart_race(
    df=df_pivot,
    filename=f"lightweight_rank_{DATE_TIME_TAG}.gif",
    orientation="h",
    sort="desc",
    n_bars=15,
    fixed_max=True,
    steps_per_period=10,
    figsize=(10, 6),
    dpi=144,
    cmap="dark12",
    title="Lightweight Ranking",
    label_bars=True,
    bar_size=0.95,
    period_label={"x": 0.99, "y": 0.25, "ha": "right", "va": "center"},
    period_fmt="%B %d, %Y",
    period_summary_func=lambda v, r: {
        "x": 0.99,
        "y": 0.18,
        "s": f"temp: {v.nlargest().sum():,.0f}",
        "ha": "right",
        "size": 8,
        "family": "Courier New",
    },
    title_size="",
    bar_label_size=7,
    tick_label_size=7,
    # shared_fontdict={"family": "Helvetica", "color": ".1"},
    scale="linear",
    writer=None,
    fig=None,
    bar_kwargs={"alpha": 0.7},
    filter_column_colors=False,
)
```
