```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pandas as pd
import seaborn as sn
import bar_chart_race as bcr
```

```python
df = pd.read_csv("monthly-temperature-anomalies.csv")
df["date"] = pd.to_datetime(df["Day"])
df = df[
    (df["date"] > "2020-01-01")
    & (
        df["Entity"].isin(
            [
                "Afghanistan",
                "Albania",
                "Algeria",
                "American Samoa",
                "Andorra",
                "Angola",
                "Anguilla",
                "Antigua and Barbuda",
                "Argentina",
                "Armenia",
                "Australia",
                "Austria",
                "Azerbaijan",
                "Bahamas",
                "Bahrain",
                "Bangladesh",
                "Belarus",
                "Belgium",
                "Belize",
                "Benin",
            ]
        )
    )
]
# df.set_index('date', inplace=True)
df.head()
```

```python
df_pivot = df.pivot(index="date", columns="Entity", values="Temperature anomaly")
df_pivot.head()
```

```python
df_pivot = df_pivot.ffill()
```

```python
df_values, df_ranks = bcr.prepare_wide_data(
    df_pivot, steps_per_period=4, orientation="h", sort="desc"
)
df_values.head()
```

```python
bcr.bar_chart_race(
    df=df_pivot,
    filename="test_6.gif",
    orientation="h",
    sort="desc",
    n_bars=10,
    fixed_max=True,
    steps_per_period=40,
    figsize=(10, 6),
    dpi=144,
    cmap="dark12",
    title="Temperature Anomaly by Country",
    label_bars=True,
    bar_size=0.95,
    period_label={"x": 0.99, "y": 0.25, "ha": "right", "va": "center"},
    period_fmt="%B %d, %Y",
    period_summary_func=lambda v, r: {
        "x": 0.99,
        "y": 0.18,
        "s": f"temp: {v.nlargest().sum():,.0f}",
        "ha": "right",
        "size": 8,
        "family": "Courier New",
    },
    title_size="",
    bar_label_size=7,
    tick_label_size=7,
    # shared_fontdict={"family": "Helvetica", "color": ".1"},
    scale="linear",
    writer=None,
    fig=None,
    bar_kwargs={"alpha": 0.7},
    filter_column_colors=False,
)
```
